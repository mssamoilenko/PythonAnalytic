```python
# Завдання 1: Робота з NumPy

# Умова завдання:

# 1 Задати одновимірний масив з 20 елементів за допомогою random().

# 2 Перетворити його у двовимірний масив (наприклад, 4x5 або 5x4).

# 3 Кожен елемент масиву збільшити на 10.

# 4 Вивести результат у файл та на екран.

# 5 Порівняти час виконання операцій з використанням NumPy та без нього.
```

```python
import numpy as np
import time
import random  # Додаємо модуль random для чистого Python
```

```python
# Великий масив для наочності
size = 100000

# NumPy
start_time_np = time.time()
array_1d_np = np.random.rand(size)
array_2d_np = array_1d_np.reshape(1000, 100)  # 1000x100
array_2d_np += 10
end_time_np = time.time()

print("Час з NumPy:", end_time_np - start_time_np, "секунд")
```

```python
# Чистий Python (без NumPy)
start_time_py = time.time()
array_1d_py = [random.random() for _ in range(size)]  # Використовуємо random.random()
array_2d_py = []
for i in range(1000):
    row = []
    for j in range(100):
        row.append(array_1d_py[i * 100 + j] + 10)
    array_2d_py.append(row)
end_time_py = time.time()

print("Час без NumPy:", end_time_py - start_time_py, "секунд")
```

```python
# Завдання 2:

# Візуалізація за допомогою Matplotlib

# Умова завдання: Побудувати графіки функцій:

# Y ( x ) = 2 ⋅ sin ⁡ ( 10 x ) Y(x)=2⋅sin(10x), де x ∈ [ − 3 , 3 ] x∈[−3,3]

# Y ( x ) = x ⋅ sin ⁡ ( 10 x ) Y(x)= x​⋅sin(10x), де x ∈ [ 0 , 5 ] x∈[0,5]

# Вимоги:

# Побудувати обидва графіки в одному вікні.

# Побудувати графіки на різних полях фігури.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 500)  # x від 0 до 5 (500 точок)
y = np.sqrt(x) * np.sin(10 * x)  # Правильна формула: sqrt(x)*sin(10x)
```

```python
# Графік у одному вікні
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=r'$Y(x) = \sqrt{x} \cdot \sin(10x)$', color='red')
plt.title("Графік функції")
plt.xlabel("x")
plt.ylabel("Y(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Графік на окремому полі (якщо потрібно порівняти з іншою функцією)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# Перший графік (інша функція, якщо потрібно)
x1 = np.linspace(-3, 3, 500)
y1 = 2 * np.sin(10 * x1)
axes[0].plot(x1, y1, color='blue')
axes[0].set_title(r'$Y(x) = 2 \cdot \sin(10x)$')

# Другий графік (наша виправлена функція)
axes[1].plot(x, y, color='red')
axes[1].set_title(r'$Y(x) = \sqrt{x} \cdot \sin(10x)$')

plt.tight_layout()
plt.show()
```
