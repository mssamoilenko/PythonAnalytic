**Завдання 1: Робота з NumPy**

Умова завдання:

1 Задати одновимірний масив з 20 елементів за
  допомогою random().

2 Перетворити його у двовимірний масив
  (наприклад, 4x5 або 5x4).

3 Кожен елемент масиву збільшити на 10.

4 Вивести результат у файл та на екран.

5 Порівняти час виконання операцій з
  використанням NumPy та без нього.

```python
import random
import numpy as np
import time

array_1d = [random.random() for _ in range(20)]

start_time_py = time.time()

rows, cols = 4, 5
array_2d_py = []
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(array_1d[i * cols + j] + 10)
    array_2d_py.append(row)

end_time_py = time.time()
time_py = end_time_py - start_time_py

start_time_np = time.time()

array_np = np.array(array_1d).reshape(rows, cols)
array_np += 10

end_time_np = time.time()
time_np = end_time_np - start_time_np

print("Результат без NumPy:")
for row in array_2d_py:
    print(row)

print("\nРезультат з NumPy:")
print(array_np)

print(f"\nЧас без NumPy: {time_py:.6f} секунд")
print(f"Час з NumPy: {time_np:.6f} секунд")

with open("output.txt", "w") as f:
    f.write("Результат без NumPy:\n")
    for row in array_2d_py:
        f.write(" ".join(map(str, row)) + "\n")

    f.write("\nРезультат з NumPy:\n")
    f.write(np.array2string(array_np))

    f.write(f"\n\nЧас без NumPy: {time_py:.6f} секунд\n")
    f.write(f"Час з NumPy: {time_np:.6f} секунд\n")

print("\nРезультати записані у файл 'output.txt'")
```

**Завдання 2:**

Візуалізація за допомогою Matplotlib

Умова завдання:
Побудувати графіки функцій:

Y ( x ) = 2^x * sin ( 10 x ), де x = [ − 3 , 3 ]

Y ( x ) = x ^(1/2)* sin ( 10 x ), де x = [ 0 , 5 ]


Вимоги:

Побудувати обидва графіки в одному вікні.

Побудувати графіки на різних полях фігури.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

x1 = np.linspace(-3, 3, 500)
y1 = (2 ** x1) * np.sin(10 * x1)

ax1.plot(x1, y1, 'b-', linewidth=2)
ax1.set_title(r'$Y(x) = 2^x \cdot \sin(10x)$ для $x \in [-3, 3]$')
ax1.set_xlabel('x')
ax1.set_ylabel('Y(x)')
ax1.grid(True)

x2 = np.linspace(0, 5, 500)
y2 = np.sqrt(x2) * np.sin(10 * x2)

ax2.plot(x2, y2, 'r-', linewidth=2)
ax2.set_title(r'$Y(x) = \sqrt{x} \cdot \sin(10x)$ для $x \in [0, 5]$')
ax2.set_xlabel('x')
ax2.set_ylabel('Y(x)')
ax2.grid(True)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x1, y1, 'b-', linewidth=2, label=r'$2^x \cdot \sin(10x)$')
plt.plot(x2, y2, 'r-', linewidth=2, label=r'$\sqrt{x} \cdot \sin(10x)$')

plt.title('Порівняння функцій')
plt.xlabel('x')
plt.ylabel('Y(x)')
plt.grid(True)
plt.legend(loc='upper right')

plt.xlim(-3, 5)
plt.ylim(-10, 10)

plt.show()
```
