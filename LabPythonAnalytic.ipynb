**Завдання 1: Робота з NumPy**

Умова завдання:

1 Задати одновимірний масив з 20 елементів за
  допомогою random().

2 Перетворити його у двовимірний масив
  (наприклад, 4x5 або 5x4).

3 Кожен елемент масиву збільшити на 10.

4 Вивести результат у файл та на екран.

5 Порівняти час виконання операцій з
  використанням NumPy та без нього.

```python
import numpy as np
import time
import random  # Додаємо модуль random для чистого Python
```

```python
# Великий масив для наочності
size = 100000

# NumPy
start_time_np = time.time()
array_1d_np = np.random.rand(size)
array_2d_np = array_1d_np.reshape(1000, 100)  # 1000x100
array_2d_np += 10
end_time_np = time.time()

print("Час з NumPy:", end_time_np - start_time_np, "секунд")
```

```python
# Чистий Python (без NumPy)
start_time_py = time.time()
array_1d_py = [random.random() for _ in range(size)]  # Використовуємо random.random()
array_2d_py = []
for i in range(1000):
    row = []
    for j in range(100):
        row.append(array_1d_py[i * 100 + j] + 10)
    array_2d_py.append(row)
end_time_py = time.time()

print("Час без NumPy:", end_time_py - start_time_py, "секунд")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Функції
x1 = np.linspace(-3, 3, 500)
y1 = 2 * np.sin(10 * x1)

x2 = np.linspace(0, 5, 500)
y2 = np.sqrt(x2) * np.sin(10 * x2)
```

```python
# Візуалізація в одному вікні
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, label='Y(x) = 2*sin(10x)', color='blue')
plt.plot(x2, y2, label='Y(x) = sqrt(x)*sin(10x)', color='red')
plt.title("Графіки функцій")
plt.xlabel("x")
plt.ylabel("Y(x)")
plt.grid()
plt.legend()
plt.show()
```

```python
# Візуалізація на різних полях
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(x1, y1, color='blue')
ax1.set_title("Y(x) = 2*sin(10x)")
ax1.grid()

ax2.plot(x2, y2, color='red')
ax2.set_title("Y(x) = sqrt(x)*sin(10x)")
ax2.grid()


plt.tight_layout()
plt.show()
```
